# Logistic Regression for Multiclass Regresssion

Predict with a multinomial and OvR Logistic Regression model

```python
# create x and y data
from sklearn.datasets import make_classification

x, y = make_classification(n_samples= 1000, n_features=10, n_informative=5, n_redundant=5, n_classes=3, random_state=1)
```

```python
y
```

```python
x
```

```python
# train test the model
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.20, random_state =42 )
```

# Define Logistic Regression model for multiclass classification

```python
from sklearn.linear_model import LogisticRegression
```

# 1. Multinomial LR model

```python
multinomial_model = LogisticRegression(multi_class = 'multinomial', solver = 'lbfgs')
```

```python
x_train.shape, x_test.shape
```

```python
# Training a the model
multinomial_model.fit(x_train, y_train)
```

```python
y_pred = multinomial_model.predict(x_test)
```

```python
y_pred
```

```python
multinomial_model.predict_proba(x_test)
```

# Performance metrics

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

print("Confusion matrix -> \n", confusion_matrix(y_test, y_pred))

print("Accuracy score -> ", accuracy_score(y_test, y_pred))

print("Classification report -> \n", classification_report(y_test, y_pred))
```

# 2. OvR LR model

```python
ovr_model = LogisticRegression(multi_class = 'ovr', solver = 'lbfgs')
```

```python
# train test split already done in first technique.

# training model
ovr_model.fit(x_train, y_train)
```

```python
y_pred = ovr_model.predict(x_test)
```

```python
y_pred
```

```python
ovr_model.predict_proba(x_test)
```

```python
# Preformance metrics for OvR model.

print("Confusion matrix -> \n", confusion_matrix(y_test, y_pred))

print("Accuracy score -> ", accuracy_score(y_test, y_pred))

print("Classification report -> \n", classification_report(y_test, y_pred))
```

# Internal Assignment
Apply mutinomial / OvR LR model in Iris Dataset

```python
from sklearn.datasets import load_iris
```

```python
dataset = load_iris()
```

```python
dataset.data
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(dataset.data, columns = dataset.feature_names)
```

```python
df.head()
```

```python
df['target'] = dataset.target
```

```python
df.tail(2)
```

```python
df['target'].value_counts()
```

As we can see there are **3 classes (0, 1 and 2)**. This is Multiclass Classification problem.

```python
x = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
x, y
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size = 0.20, random_state = 42)
```

```python
X_train.shape, X_test.shape
```

```python
# Define model

model = LogisticRegression(multi_class = 'multinomial', solver = 'lbfgs')
```

```python
# train the model

model.fit(X_train, Y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
model.predict_proba(X_test)
```

```python
# Performance metrics


print("Confusion matrix -> \n", confusion_matrix(Y_test, y_pred))

print("Accuracy score -> ", accuracy_score(Y_test, y_pred))

print("Classification report -> \n", classification_report(Y_test, y_pred))
```
