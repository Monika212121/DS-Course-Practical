# Anamoly detection by DBSCAN Clustering

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from sklearn.cluster import DBSCAN
```

```python
# Creating non-linear cluster

x, y = make_circles(n_samples = 800, factor = 0.2, noise = 0.1)
```

```python
x
```

```python
# Plotting graph for x1 vs x2

plt.scatter(x[:,0], x[:,1])
```

```python
plt.scatter(x[:,0], x[:,1], c = y)
```

## Implementation DBSCAN

```python
dbscan = DBSCAN(eps = 0.1)
dbscan
```

```python
y_pred = dbscan.fit_predict(x)
```

```python
y_pred
```

```python
# both y_pred and dbscan.labels_ are same, have same values

plt.scatter(y_pred, dbscan.labels_)
```

```python
plt.scatter(x[:,0], x[:,1], c = y_pred)
```

# To find outliers

```python
# getting index of outliers (y_pred = -1)

index = np.where(y_pred < 0)
index
```

```python
plt.scatter(x[:,0], x[:,1])                                 # normal graph

plt.scatter(x[index,0], x[index,1], edgecolors = 'r')       # presents outliers
```
