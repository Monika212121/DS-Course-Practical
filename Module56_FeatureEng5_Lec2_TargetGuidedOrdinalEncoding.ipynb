TARGET GUIDED ORDINAL ENCODING

```python
import pandas as pd

df = pd.DataFrame({
    'city': ['Ranchi', 'Bangalore', 'Mumbai', 'Pune', 'Bangalore', 'Ranchi'],
    'price': [100, 900, 850, 400, 1000, 120]
})
```

```python
df
```

NOTE: If there is any outlier, then take median , if not take mean.

```python
# Calculate the mean price for each city

# similar to SQL
df.groupby('city')['price']
```

```python
df.groupby('city')['price'].mean()
```

```python
# Convert the above in a dictionary

mean_price = df.groupby('city')['price'].mean().to_dict()
```

```python
mean_price
```

```python
# Replace each city with its mean price

df['city_encoded'] = df['city'].map(mean_price)
```
