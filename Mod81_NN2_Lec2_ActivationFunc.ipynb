# Activation functions

````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
# Defininng weights
w = [1, 2]
b = 5

# Create data
x = np.linspace(-5, 5, 100)
y = w[0]*x + w[1]*x + b
````

````python
# Plot linear combination
plt.plot(x, y, label = 'Linear Combination')
plt.legend()
plt.show()
````

# Activation function directly using their formulas

## 1.) Sigmoid Activation function

````python
sigmoid  = lambda x: 1/(1+np.exp(-x))
y_sigmoid = sigmoid(y)
````

````python
# Plotting X vs Sigmoid function values

plt.plot(x, y_sigmoid, label = 'Sigmoid Activation function')
plt.legend()
plt.show()
````

## 2.) ReLU Activation Function

````python
relu = lambda x: np.maximum(x, 0)
y_relu = relu(y)
plt.plot(x, y_relu, label = 'ReLU Activation Function')

plt.legend()
plt.xlabel('x')
plt.ylabel('y')

plt.show()
````

# Generic function to plot Activation function

````python
import seaborn as sns
import tensorflow as tf
plt.style.use('fivethirtyeight')
````

````python
x = np.linspace(-10, 10, 100)

def plot_this_graph(x, y, ALPHA = 0.6, label_x = r"$x \rightarrow$", label_y = r"$act(x) \rightarrow$", title = None, LABEL = None):

  plt.figure(figsize = (6,5))
  plt.axhline(y= 0, color= "black", linestyle= "--", lw= 2)
  plt.axvline(x= 0, color= 'black', linestyle= "--", lw= 2)
  plt.xlabel(label_x)
  plt.ylabel(label_y)
  plt.title(title)

  if LABEL != None:
    plt.plot(x, y, alpha = ALPHA, label= LABEL)
    plt.legend(fontsize = 12)

  else:
    plt.plot(x, y, alpha = ALPHA)
````

## To find derivative of a function

````python
def derivative_func(f, z, delta_z= 1e-6):
  return ( f(z+delta_z) - f(z) ) / (delta_z)     # f'(x) = [f(b)-f(a)] / (b-a)
````

# Activation function using module "Tensorflow"

## 1.) Sigmoid function

````python
def sigmoid(x):
  return tf.keras.activations.sigmoid(x)
````

````python
plot_this_graph(x, sigmoid(x), title = "Sigmoid Function", LABEL= r"$\sigma(x) = \frac{1}{1 + e^{-x}} $")

plot_this_graph(x, derivative_func(sigmoid, x), title = "Derivative of Sigmoid function", LABEL = r"$\sigma^\prime(x) = \sigma(x)(1-\sigma(x))$")
````

## 2.) ReLU function

````python
def relu(x):
  return tf.keras.activations.relu(x, negative_slope = 0.0, max_value= None, threshold = 0)

# NOTE: Here, negative_sope is alpha value, 'alpha' is degraded in latest version.
````

````python
plot_this_graph(x, relu(x), title = "ReLU", LABEL = r"$ReLU(x) = max(x,0)$")

plot_this_graph(x, derivative_func(relu, x), title = "Derivative of ReLU function", LABEL = r"$ReLU^\prime(x)$")
````

## 3.) Leaky ReLU function

````python
def leaky_relu(x, alpha = 0.3):
  return np.where(x >= 0, x, alpha*x)
````

````python
plot_this_graph(x, leaky_relu(x), title = "Leaky ReLU")

plot_this_graph(x, derivative_func(leaky_relu, x), title = "Derivative of LeakyReLU function", LABEL = "$Leaky\_ReLU^\prime(X)$")
````

## 4.) Softmax function

````python
def softmax(x):
  return tf.keras.activations.softmax(tf.constant([[0.3, 0.3, 0.4, 0.5, 0.7]]), axis = -1)
````

## 5.) Hard Sigmoid
         
```
hard_sigmoid  = 0                , x < -2.5
              = 1                , x > 2.5
              = 0.2 * x + 0.5    , -2.5 <= x <= 2.5
```



````python
def hard_sigmoid(z):
  z = tf.constant(z)
  return tf.keras.activations.hard_sigmoid(z)
````

````python
plot_this_graph(x, hard_sigmoid(x), title = "Hard Sigmoid", LABEL = r"$hard\_sigmoid(x)$")

plot_this_graph(x, derivative_func(hard_sigmoid, x), title = "Derivative of Hard Sigmoid", LABEL = "$hard\_sigmoid^\prime(x)$")
````

Hard Sigmoid function saves time but lacks accuracy.
