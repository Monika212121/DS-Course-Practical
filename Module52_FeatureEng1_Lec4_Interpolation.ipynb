DATA INTERPOLATION

```python
import numpy as np

x= np.array([1,2,3,4,5])
y= np.array([10,20,30,40,50])
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

LINEAR INTERPOLATION using NumPy :

```python
# Interpolate the datapoints using:

x_new = np.linspace(1,5,10)  # create new x values
y_inter = np.interp(x_new, x, y) # interpolate y values
```

```python
x_new
```

```python
print(y_inter)
```

```python
plt.scatter(x_new, y_inter)
```

CUBIC INTERPOLATION with SciPy :

```python
x1= np.array([1,2,3,4,5])
y1= np.array([1,8,27,64,125])
```

```python
plt.scatter(x1,y1)
```

```python
from scipy.interpolate import interp1d
```

```python
# Creating a cubic interpolation function

f = interp1d(x1,y1,kind='cubic')
```

```python
f
```

```python
x1_new = np.linspace(1,5,10)
y1_inter = f(x1_new)
```

```python
x1_new
```

```python
y1_inter
```

```python
plt.scatter(x1_new, y1_inter)
```
