```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()

# bokeh.io.output_notebook() enables rendering of graphs in this notebook.
# If not present ,then we can't see any errors or output graphs in console.
```

```python
from bokeh.plotting import figure, output_file, show
from bokeh.sampledata.iris import flowers
flowers
output_file('test.html')

p = figure(title = "test flower")
p.xaxis.axis_label = "petal length"
p.yaxis.axis_label = "petal width"
p.circle(flowers['petal_length'], flowers['petal_width'])
show(p)
```

```python
from bokeh.plotting import figure, output_file, show
from bokeh.sampledata.iris import flowers
flowers
output_file('line.html')

p = figure(title = "test flower")
p.xaxis.axis_label = "x axis"
p.yaxis.axis_label = "y axis"
p.circle(flowers['petal_length'], flowers['petal_width'])
show(p)
```

```python
x = [2,3,4,5,6]
y= [3,4,5,6,7]

output_file("line.html")

p = figure(title = 'line plot')
p.line(x,y)
show(p)
```

```python
x = [2,3,4,5,6]
y= [3,4,5,6,7]

output_file("line.html")

p = figure(title = 'line plot')
p.scatter(x,y)
show(p)
```

```python
x = [2,3,4,5,6]
y= [3,4,5,6,7]

output_file("line.html")

p = figure(title = 'line plot')
p.scatter(x,y, color = 'red')
show(p)
```

```python
x = [2,3,4,5,6]
y= [3,4,5,6,7]

output_file("line.html")

p = figure(title = 'line plot')
p.scatter(x,y, fill_color = 'red', legend_label = 'red points')
show(p)
```

```python
x = [2,3,4,5,6]
y= [3,4,5,6,7]

output_file("line.html")

p = figure(title = 'line plot')
p.scatter(x,y, fill_color = 'red', legend_label = 'red points', size = 20 )
show(p)
```
