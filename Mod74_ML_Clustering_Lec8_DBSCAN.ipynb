# DBSCAN Clustering Implementation

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Create Non linear cluster data using make_moons

from sklearn.datasets import make_moons
```

```python
x, y = make_moons(n_samples = 500, noise = 0.10)
```

```python
x
```

```python
plt.scatter(x[:,0], x[:,1])
```

# Standarization

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
x_scaled = scaler.fit_transform(x)
```

# DBSCAN Implementation

```python
from sklearn.cluster import DBSCAN
```

```python
dbscan = DBSCAN(eps = 0.3)
```

```python
dbscan.fit(x_scaled)
```

```python
dbscan.labels_
```

```python
plt.scatter(x[:,0], x[:,1], c = dbscan.labels_)
```

```python
# For checking the accuracy

plt.scatter(x[:,0], x[:,1], c = y)
```

# Analysis:

We can see here that the predicted and the actiual y values generate same clusters.

Hence, the clustering is accurate.
