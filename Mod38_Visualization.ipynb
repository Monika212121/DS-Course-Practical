```python
import seaborn as sns
```

```python
iris= sns.load_dataset('iris')
```

```python
iris
```

```python
# If there is any relationship between sepal_width and petal width
```

```python
sns.scatterplot(x = iris.sepal_length, y = iris.sepal_width)
```

```python
# As we can see there are 2 spaces where we can see linear relationship between the 2 variables, but overall there is not pattern.
```

```python
sns.scatterplot(x = iris.sepal_length, y = iris.petal_width)
```

```python
# We can see that for some region(x< 6) , petal_width is constant with the sepal_length and region (x > 6), the petal_width increases linearly w.r.t sepal_length.
```

```python
sns.scatterplot(x = iris.petal_width, y = iris.petal_length)
```

```python
sns.displot(iris['sepal_length'])
```

RELATIONAL PLOT:

```python
sns.relplot(x = tips.total_bill, y= tips.tip, data = tips, hue = 'smoker')
```

```python
sns.relplot(x = tips.total_bill, y= tips.tip, data = tips, hue = 'size')
```

```python
tips['size'].value_counts()
```

```python
sns.relplot(x = tips.total_bill, y= tips.tip, data = tips, style = 'size')
```

```python
# style used for grouping with diff symbols and hue differentiates w.r.t colour shade.
```

```python
sns.relplot(x = tips.total_bill, y= tips.tip, data = tips, style = 'size',  hue = 'time')
```

```python
# Here I am separating data w.r.t. 2 parameters, first w.r.t size using 'style' and second w.r.t. time using 'hue'.
```

CATEGORICAL PLOT:

```python
sns.catplot(x= 'day', y= 'total_bill', data= tips)
```

```python
sns.catplot(x= 'smoker', y= 'tip', data= tips)
```

JOINT PLOT:

```python
sns.jointplot(x=tips.total_bill, y= tips.tip)
```

```python
# It joins scatterplot + histogram in one graph.
```
