SMOTE(Synthetic Minority Oversampling Technique)

```python
from sklearn.datasets import make_classification
```

```python
X, Y = make_classification(n_samples= 1000, n_features=2, n_clusters_per_class=1, n_redundant=0, weights=[0.90], random_state=1)
```

```python
# dependent feature

Y
```

TO CREATE DATASET FROM THIS VARIABLES DATA

```python
import pandas as pd

df1 = pd.DataFrame(X, columns = ['f1', 'f2'])
df2 = pd.DataFrame(Y, columns = ['target'])
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1,df2])
```

```python
# concatanating column wise

final_df = pd.concat([df1,df2], axis =1)
```

```python
final_df.head()
```

```python
final_df['target'].value_counts()
```

```python
# Plot this dataset

import matplotlib.pyplot as plt
plt.scatter(final_df['f1'], final_df['f2'], c=final_df['target'])
```

APPLYING SMOTE IN THIS DATASET

```python
!pip install imblearn
```

```python
from imblearn.over_sampling import SMOTE
```

```python
# TRANSFORM THE DATASET
# Initializing SMOTE

oversample = SMOTE()
X1, Y1 = oversample.fit_resample(final_df[['f1','f2']], final_df['target'])
```
