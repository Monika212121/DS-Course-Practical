DATA ENCODING

NOMINAL / ONE HOT ENCODING TECHNIQUE :

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
import pandas as pd
```

```python
df= pd.DataFrame({
    'color': ['red', 'blue', 'green', 'red', 'white', 'blue', 'black']
})
```

```python
df
```

```python
# Create an instance of One Hot Enncoder
encoder = OneHotEncoder()
```

```python
encoder
```

```python
# fit the encoder to the dataframe and transform the categorical variable

encoded_data = encoder.fit_transform(df[['color']])
```

```python
encoder.fit_transform(df[['color']]).toarray()
```

CREATE DATAFRAME FROM THIS DATA:

```python
encoder.get_feature_names_out()
```

```python
encoded_df = pd.DataFrame(encoded_data.toarray(), columns = encoder.get_feature_names_out())
```

```python
encoded_df
```

```python
# concatenate both dataframe:

pd.concat([df, encoded_df], axis=1)
```
