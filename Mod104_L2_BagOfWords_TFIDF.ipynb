# Preprocessing of data

```python
import re
from nltk.corpus import stopwords
nltk.download('stopwords')
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer
```

```python
# cleaning data
sentences = nltk.sent_tokenize(paragraph)
stemmer = PorterStemmer()
corpus = []

for i in range(len(sentences)):
  sent = re.sub('[^a-zA-Z]', ' ', sentences[i])
  sent = sent.lower()
  sent = sent.split()

  filtered_words = [stemmer.stem(word) for word in sent if word not in set(stopwords.words('english'))]
  filtered_words = ' '.join(filtered_words)
  corpus.append(filtered_words)
```

# Bag of Words implementation

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 1000)
X =  cv.fit_transform(corpus).toarray()
X
```

# TF- IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(corpus).toarray()
X
```

```python
import pandas as pd
```

```python
df = pd.read_csv('topicsData.csv', encoding = "ISO-8859-1")
df.head()
```

```python
df.Date.unique()
```

```python
# Splitting train and test data, also splitting x and y for both train and test

train = df[df['Date'] < '20150101']
test = df[df['Date'] > '20141231']

train_y = train['Label']
test_y = test['Label']
```

```python
train.shape, test.shape
```

```python
# removing punctuation
data = train.iloc[:,2:27]         # removes first 2 columns 'date' and 'label'
data.replace('[^a-zA-Z]', " ", regex = True, inplace = True)

# renaming columns for easy access
list1 = [i for i in range(25)]
new_indices = [str(ind) for ind in list1]

data.columns = new_indices
data.head()
```

```python
df.columns
```

```python
data.columns
```

```python
# converting to lowercase values
for ind in new_indices:
  data[ind] = data[ind].str.lower()

data.head(2)
```

```python
' '.join(str(x) for x in data.iloc[1, 0:25])        # contains contents of second row(index = 1)
```

```python
headlines = []
for row in range(0, len(data.index)):
  headlines.append(' '.join(str(x) for x in data.iloc[row, 0:25]))
```

```python
headlines[0]
```

```python
len(headlines)
```

```python
headlines[32]
```

```python
# Applying Bag Of Words
cv = CountVectorizer(max_features = 1500, ngram_range = (2, 2))
train_data = cv.fit_transform(headlines)
```

```python
print(train_data[0].toarray())

print(train_data[0].shape)
```

```python
# Preprocessign test data
test_transform = []

for row in range(0, len(test.index)):
  test_transform.append(' '.join(str(x) for x in test.iloc[row, 0:25]))

test_data = cv.fit_transform(test_transform)
```

```python
print(test_data.shape)
```

# Implementing ML algorithm

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
model = RandomForestClassifier(n_estimators= 200, criterion = 'entropy')
model.fit(train_data, train_y)
```

```python
y_pred = model.predict(test_data)
```

```python
y_pred
```

## Performance metrics

```python
cm = confusion_matrix(test_y, y_pred)
ac_sc = accuracy_score(test_y, y_pred)
cl_report = classification_report(test_y, y_pred)

print("Confusion matrix: \n", cm)
print("\nAccuracy score: ", ac_sc)
print("\nClassification report: \n", cl_report)
```

# Analysis:

`Use case :` ML  problem predciting 'Label', present as first column in the original datast.

- What we did here is we took a dataset, it contains a lot of test data. So we needed to convert them to numbers(vectors) so that we can train the text data into any ML Model for predicting labels for them.

- Step1: We converted text data -> vectors/numbers using BoW model.

    - Here, we are converting train and test data into vecotrs separately.

- Step2: Then we are using this vector dataset instead of the original text dataset in ML algorithm `RandomForestClassifier()` to predict the label.
