Imbalanced Dataset Handling :-

1. Upsampling
2. Downsampling

```python
import numpy as np
import pandas as pd
```

```python
# Set the random dataset for reproducibility
np.random.seed(123)

# Create a dataframe with 2 classes
n_samples = 1000
class_0_ratio = 0.9
n_class_0 = int(n_samples * class_0_ratio)
n_class_1 = n_samples - n_class_0
```

```python
n_class_0, n_class_1
```

```python
# Creating dataframe:

class_0 = pd.DataFrame({
    'feature_1' : np.random.normal(loc=0, scale=1, size=n_class_0),
    'feature_2' : np.random.normal(loc=0, scale=1, size=n_class_0),
    'target': [0] * n_class_0
})

class_1 = pd.DataFrame({
    'feature_1' : np.random.normal(loc=2, scale=1, size=n_class_1),
    'feature_2' : np.random.normal(loc=2, scale=1, size=n_class_1),
    'target': [1] * n_class_1
})

# It will create dataset containing 900 0's and 100 1's.
```

```python
class_1
```

```python
pd.concat([class_0, class_1])
```

```python
df = pd.concat([class_0, class_1]).reset_index(drop=True)
```

```python
df.head()
```

```python
df['target'].value_counts()
```

UPSAMPLING

```python
df_minority = df[df['target'] == 1]
df_majority = df[df['target'] == 0]
```

```python
df_minority.head()
```

```python
df_majority.head()
```

```python
# Perform Upsampling

from sklearn.utils import resample
```

```python
df_minority_upsample = resample(
    df_minority,
    replace = True, # Sample with replacement
    n_samples= len(df_majority), # to match the majority class
    random_state = 45
)
```

```python
df_minority_upsample.shape
```

```python
df_minority.shape
```

```python
df_minority_upsample.head()
```

```python
df_minority_upsample.value_counts()
```

```python
df_minority_upsample['target'].value_counts()
```

```python
df_upsampled = pd.concat([df_majority, df_minority_upsample])
```

```python
df_upsampled.shape
```

```python
df_upsampled['target'].value_counts()
```

```python
df_upsampled.head()
```

```python
df_upsampled.tail()
```

DOWNSAMPLING:

```python
class_0 = pd.DataFrame({
    'feature_1' : np.random.normal(loc=0, scale=1, size=n_class_0),
    'feature_2' : np.random.normal(loc=0, scale=1, size=n_class_0),
    'target': [0] * n_class_0
})

class_1 = pd.DataFrame({
    'feature_1' : np.random.normal(loc=2, scale=1, size=n_class_1),
    'feature_2' : np.random.normal(loc=2, scale=1, size=n_class_1),
    'target': [1] * n_class_1
})
```

```python
df = pd.concat([class_0, class_1]).reset_index(drop=True)
```

```python
df_majority_downsample = resample(
    df_majority,
    replace = False, # Sample without replacement
    n_samples= len(df_minority), # to match the minority class
    random_state = 45
)
```

```python
df_majority_downsample.shape
```

```python
df_downsample = pd.concat([df_minority, df_majority_downsample])
```
