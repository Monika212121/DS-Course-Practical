# Local Outlier Factor(LOF) for Anomaly detection

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from sklearn.cluster import DBSCAN
```

```python
# Creating non-linear cluster

x, y = make_circles(n_samples = 800, factor = 0.2, noise = 0.1)
```

```python
# Plotting graph for x1 vs x2

plt.scatter(x[:,0], x[:,1])
```

# LocalOutlier Implementation

```python
from sklearn.neighbors import LocalOutlierFactor
```

```python
lof = LocalOutlierFactor(n_neighbors = 5)
```

```python
y_pred = lof.fit_predict(x)
```

```python
y_pred
```

```python
index = np.where(y_pred < 0)
index
```

```python
# Plotting graph for x1 vs x2

plt.scatter(x[:,0], x[:,1], c = y_pred)

plt.scatter(x[index,0], x[index,1], edgecolors = 'r')
```

## Analysis:

After comparing from DBSCAN clustring graph, the outliers here are different than the outlier in DBSCAN.

It might be because the outliers here are local whereas in DBSCAN the outliers are Golbal as well.

# NOTE:

The outliers here are local outliers.

It uses K- Means algorithmn to check density of the evaluating point and its 'k' neighbors.

If the density is less than its neighbors , then the particular point is the Local outlier.
